```python
def print_key_tree(data, indent=0):
    prefix = "  " * indent + "- "

    if isinstance(data, dict):
        for k, v in data.items():
            print(f"{prefix}{k}")
            print_key_tree(v, indent + 1)

    elif isinstance(data, list) and data:
        print(prefix + "[list]")
        print_key_tree(data[0], indent + 1)

    elif isinstance(data, list):
        print(prefix + "[empty list]")

# use it
print_key_tree(data)
```
